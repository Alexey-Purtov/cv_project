```python
from ultralytics import YOLO
import torch
```

```python
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
DEVICE
```

```python
model = YOLO("yolo11m.pt")
```

```python
results = model.train(
    data=r"D:\ds\ds-phase-2-master\cvProject\face.yaml",
    epochs=10,
    imgsz=640,
    batch=16,
    device=DEVICE
)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

epochs = list(range(1, 11))

# Значения метрик
map_50 = [0.74, 0.755, 0.787, 0.817, 0.825, 0.845, 0.88, 0.866, 0.873, 0.88]
map_5095 = [0.446, 0.472, 0.497, 0.515, 0.531, 0.549, 0.554, 0.572, 0.578, 0.587]
box_loss = [1.427, 1.411, 1.341, 1.381, 1.26, 1.244, 1.2, 1.174, 1.154, 1.131]
cls_loss = [1.15, 0.992, 0.899, 0.843, 0.773, 0.721, 0.687, 0.659, 0.624, 0.586]
dfl_loss = [1.285, 1.128, 1.26, 1.233, 1.073, 1.177, 1.159, 1.142, 1.133, 1.118]

plt.figure(figsize=(14, 8))

# mAP график
plt.subplot(2, 2, 1)
plt.plot(epochs, map_50, label="mAP@0.5", marker="o")
plt.plot(epochs, map_5095, label="mAP@0.5:0.95", marker="o")
plt.title("mAP по эпохам")
plt.xlabel("Эпоха")
plt.ylabel("Значение")
plt.legend()
plt.grid(True)

# Losses график
plt.subplot(2, 2, 2)
plt.plot(epochs, box_loss, label="Box Loss", marker="o")
plt.plot(epochs, cls_loss, label="Class Loss", marker="o")
plt.plot(epochs, dfl_loss, label="DFL Loss", marker="o")
plt.title("Loss по эпохам")
plt.xlabel("Эпоха")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение файла
df = pd.read_csv("runs/detect/train16/results.csv")
epochs = df.index + 1

plt.figure(figsize=(16, 10))

# mAP + Recall
plt.subplot(2, 2, 1)
plt.plot(epochs, df['metrics/mAP50(B)'], label="mAP@0.5", marker="o")
plt.plot(epochs, df['metrics/mAP50-95(B)'], label="mAP@0.5:0.95", marker="o")
plt.plot(epochs, df['metrics/recall(B)'], label="Recall", marker="o")
plt.title("mAP и Recall по эпохам")
plt.xlabel("Эпоха")
plt.ylabel("Значение")
plt.legend()
plt.grid(True)

# Precision
plt.subplot(2, 2, 2)
plt.plot(epochs, df['metrics/precision(B)'], label="Precision", color="orange", marker="o")
plt.title("Precision по эпохам")
plt.xlabel("Эпоха")
plt.grid(True)
plt.legend()

# Losses
plt.subplot(2, 2, 3)
plt.plot(epochs, df['val/box_loss'], label="Box Loss", marker="o")
plt.plot(epochs, df['val/cls_loss'], label="Cls Loss", marker="o")
plt.plot(epochs, df['val/dfl_loss'], label="DFL Loss", marker="o")
plt.title("Loss по эпохам")
plt.xlabel("Эпоха")
plt.ylabel("Loss")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

```
